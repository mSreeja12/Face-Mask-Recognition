```python
import os
import cv2

# Define the base directory where the dataset is stored
base_dir = r'face_mask_dataset\Dataset'  # Use the correct base path
data = []
categories = ['with_mask', 'without_mask']

for category in categories:
    path = os.path.join(base_dir, category)
    label = categories.index(category)
    # Check if the directory exists
    if not os.path.exists(path):
        print(f"Directory does not exist: {path}")
        continue

    for file in os.listdir(path):
        img_path = os.path.join(path, file)
        img_data = cv2.imread(img_path)
        
        if img_data is not None:
            img = cv2.resize(img_data, (128, 128))
            data.append([img,label])
            print(img.shape)
        else:
            print(f"Failed to read image: {img_path}")
```

```python
x = []
y = []

for features,label in data:
    x.append(features)
    y.append(label)
```

```python
import numpy as np
X = np.array(x)
Y = np.array(y)
```

```python
X = X/255
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from keras.applications.vgg16 import VGG16
from keras.models import Sequential
from keras.layers import Dense, Flatten

# Load the VGG16 model without the top layer (include_top=False)
vgg = VGG16(include_top=False, input_shape=(224, 224, 3))

# Create a new Sequential model
model = Sequential()

# Add VGG16 layers to the new model
for layer in vgg.layers[:-1]:
    model.add(layer)

# Optionally set the layers to be non-trainable
for layer in model.layers:
    layer.trainable = False

# Add new layers on top of the VGG16 base
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Display the model summary
model.summary()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import to_categorical
import numpy as np

num_classes = len(np.unique(Y_train))

Y_train = to_categorical(Y_train, num_classes)
Y_test = to_categorical(Y_test, num_classes)

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))  # Output layer with num_classes

model.summary()

datagen = ImageDataGenerator()

train_generator = datagen.flow(X_train, Y_train, batch_size=16)
validation_generator = datagen.flow(X_test, Y_test, batch_size=16)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(
    train_generator,
    steps_per_epoch=len(X_train) // 16,
    epochs=5,
    validation_data=validation_generator,
    validation_steps=len(X_test) // 16
)
```

```python
cap = cv2.VideoCapture(0)
```

```python
def detect_face_mask(img):
    y_pred = model.predict(img.reshape(1,128,128,3))
    
    return y_pred[0][0]
```

```python
while True:
    ret,frame = cap.read()
    frame = cv2.resize(frame,(128,128))
    y_pred = detect_face_mask(frame)
    print(y_pred)
    
    cv2.imshow("window",frame)
    
    if cv2.waitKey(1) & 0xFF ==ord('x'):
        break
        
cv2.destroyAllWindows()
```
